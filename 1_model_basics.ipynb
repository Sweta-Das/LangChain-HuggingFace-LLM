```python
import os
import torch
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer
```

### Using Tiny Llama LLM Model









```python
# Looking at tokenizer
model_name = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
tokenizer = AutoTokenizer.from_pretrained(model_name)
# print(tokenizer)
```

```python
prompt = "Who is Ada Lovelace?"
encoded_prompt = tokenizer.encode(prompt)
print(f"Prompt: {prompt}")
print(f"Encoded: {encoded_prompt}")
```

```python
for item in encoded_prompt:
  decoded = tokenizer.decode(item)
  print(f"|{item}|: |{decoded}|")
```

```python
model = AutoModelForCausalLM.from_pretrained(model_name)

generator = pipeline('text-generation', model=model, tokenizer=tokenizer)
```

```python
# Generating text based on a prompt
prompt1 = "Who is Ada Lovelace?"
answer = generator(prompt1, max_length=50, truncation=True)
answer[0]['generated_text']
```

```python
# Generating text based on a prompt
prompt2 = "What is the capital of France?"
answer = generator(prompt2, max_length=50, truncation=True)
answer[0]['generated_text']
```
